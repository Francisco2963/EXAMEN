```python
x = [' '.join(map(str,item)) for item in gb_A1.index]
y = gb_A1.values
plt.figure(figsize=(10,5))
plt.title('Comparacion de Abandonos por genero')
plt.bar(x,y)
plt.xlabel('gender')
plt.ylabel('Churn')
```
